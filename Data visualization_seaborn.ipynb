(univariate analysis --->distplot is used)
The distplot shows the distribution of a univariate set of observations.

```python
sns.distplot(data['total_bill']) #here 'kde' parameter have deafult value is True .so, we get distribution curve. 
```

x-axis show total bill and y-axis show random values.
curve line is called KDE(Kernel density estimator)

```python
sns.distplot(data['total_bill'],kde=False)
```

Bivariate analysis-->use jointplot


```python
sns.jointplot(x='total_bill',y='tip',data=data,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=data,kind='hex')#darker the hexagon more the density.
```

```python
sns.jointplot(x='total_bill',y='tip',data=data,kind='reg')
```

pairplot :it will plot pairwise rltnship across an entire for the dataframe only for numerical columns .it also supports color 
hue argument for categorical columns.

```python
sns.pairplot(data)
```

```python
sns.pairplot(data,hue='sex',palette='coolwarm')
```

```python
sns.pairplot(data,hue='smoker',palette='coolwarm')
```

rugplot:
they just draw a dashmark for every point on a univariate distribution.

```python
sns.rugplot(data['total_bill'])
```

Categoricalplot:

```python
sns.countplot(x='smoker',data=data)
```

```python
#Boxplot
sns.boxplot(x='day',y='total_bill',data=data,palette='rainbow')
```

```python
sns.boxplot(x="day",y="total_bill",hue="smoker",data=data,palette="coolwarm")
```

violinplot:

```python
sns.violinplot(x='day',y='total_bill',data=data,palette='rainbow')
```

```python
sns.boxplot(data=data,palette='rainbow',orient='h')
```

```python
sns.violinplot(x='day',y='total_bill',hue='sex',data=data,palette='rainbow')
```

```python
sns.violinplot(x='day',y='total_bill',data=data,hue='sex',split=True,palette='Set1')
```
