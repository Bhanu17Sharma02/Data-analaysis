```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv("data.csv")
```

```python
dataset.head()
```

```python
age=dataset["Age"]
alld=dataset["All_Devs"]
py=dataset["Python"]
ja=dataset["JavaScript"]
```

```python
fig,axis=plt.subplots(3,3)
```

```python
fig
```

```python
axis
```

```python
fig1,ax1=plt.subplots()
fig2,ax2=plt.subplots()
ax1.plot(age,alld,label="all_devs")
ax2.plot(age,py,label="python")
ax2.plot(age,ja,label="javascript")
ax2.legend()
plt.show()
```

```python
fig1
fig2
```

```python
fig,axis=plt.subplots(1,3,sharey=True)
dataset.plot(kind="scatter",x="Python",y="Age",ax=axis[1])
dataset.plot(kind="scatter",x="All_Devs",y="Age",ax=axis[0])
dataset.plot(kind="scatter",x="JavaScript",y="Age",ax=axis[2])
plt.show()
```
