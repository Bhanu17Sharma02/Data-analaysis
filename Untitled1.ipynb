```python
#Design for stats calculator

from tkinter import *
from scipy.stats import norm
from scipy.stats import chi2_contingency
from scipy.stats import chi2
from scipy.stats import ttest_ind


#------------------------------for Z-score--------------------- 
def z(): 
    #root.config(cursor="watch")
    root.destroy()
    def click():   #event handler
        x=int(var1.get())
        u=int(var2.get())
        v=int(var3.get())
        z=(x-u)/v
        p=norm.cdf(z)
        
        p=round(p*100,2)
        lb1_result.config(text="zvalue[{0}],p-value[{1}%]".format(z,p))
       
        
 #-------------------------z score window-------------------------------------   
    root1=Tk()
    root1.geometry("500x300")
    root1.title("Z-Score Evaluation")
    

    lb_head=Label(root1,text="Z-Score value",bg="blue",font=1)
    lb_head.grid(row=0,column=0)

    lb1_sample=Label(root1,text="Sample_Mean =",font=1,bg="gold")
    lb1_sample.config()
    lb1_sample.grid(row=1,column=0,padx=10,pady=10)
    
    var1=StringVar()
    entry_sample_mean= Entry(root1,font=1,textvariable=var1)
    entry_sample_mean.grid(row=1,column=1)
    
    lb1_pop=Label(root1,text="Population_Mean =",font=1,bg="gold")
    
    lb1_pop.grid(row=2,column=0,padx=10)
    
    var2=StringVar()
    entry_pop_mean= Entry(root1,font=1,textvariable=var2)
    entry_pop_mean.grid(row=2,column=1)
    
    
    lb1_std=Label(root1,text="Standard Deviation =",font=1,bg="gold")

    lb1_std.grid(row=3,column=0,padx=10,pady=10)
    
    var3=StringVar()
    entry_std= Entry(root1,font=1,textvariable=var3)
    entry_std.grid(row=3,column=1)
    
    btn_calc= Button(root1,text="Press for Calculate",font=1,command=click)
    btn_calc.grid(row=4,column=1)
    
    lb1_result=Label(root1,text="Result show Here",font=1,bg="green")
    lb1_result.grid(row=5,column=1,pady=20)
    
    
    root1.resizable(False,False)
    
    root1.mainloop()
#-----------------------------------ends here for z score------------------

#--------------------------------------chi starts---------------------------    
def chi():
    
    
    root.destroy()
    root1=Tk()
    root1.geometry("600x500")
    root1.title("Chi-Square Evaluation")
    
    def click():
        r=[]
        var11=var1.get()
        var22=var2.get()
        x=(list(map(int,var11.split(","))))
        y=(list(map(int,var22.split(","))))
        r.append(x)
        r.append(y)
        stats,p,dof,expected=chi2_contingency(r)
        critical=chi2.ppf(0.95,dof)
        
        lb1_result.config(text="chi2-value={0}".format(stats))
        
        if stats>critical:
            lb2_result.config(text="{0}>{1}:".format(round(stats,2),round(critical,2))+
                              "/nReject Null Hypothesis")
        else:
            lb2_result.config(text="{0}<{1}:".format(round(stats,2),round(critical,2))+
                              "Fail to Reject Null Hypothesis")
        print(r)
    
    lb_head=Label(root1,text="Chi-Square Evaluation",bg="silver",fg="blue",
                  font="Lucid  19 bold",justify="center")
    lb_head.grid(row=0,column=0,padx=5)
    
   
    lb1_sample=Label(root1,text="Data_1 =",font=1,bg="gold"
                     ,padx=20)
    lb1_sample.grid(row=1,column=0,padx=5,pady=10)
    
    var1=StringVar()
    entry_sample_mean= Entry(root1,font=1,textvariable=var1, borderwidth=5,
                             width=30,cursor="arrow")
    entry_sample_mean.grid(row=1,column=1,padx=5)
    
    lb1_sample=Label(root1,text="Insert ROW_1 elements seprated by comma",font=1,bg="silver"
                     ,fg="black"
                     ,padx=20)
    lb1_sample.grid(row=2,column=1,padx=5,pady=10)
    
    
    lb1_sample=Label(root1,text="Data_2 =",font=1,bg="gold"
                     ,padx=20)
    lb1_sample.grid(row=3,column=0,padx=5,pady=10)
    
    var2=StringVar()
    entry_sample_mean= Entry(root1,font=1,textvariable=var2, borderwidth=5,
                             width=30,cursor="arrow")
    entry_sample_mean.grid(row=3,column=1,padx=5)
    
    lb1_sample=Label(root1,text="Insert ROW_2 elements seprated by comma",font=1,bg="silver"
                     ,fg="black"
                     ,padx=20)
    lb1_sample.grid(row=5,column=1,padx=5,pady=10)
    
    btn_calc= Button(root1,text="Press for Calculate",font=1,command=click)
    btn_calc.grid(row=6,column=1)
    
    lb1_result=Label(root1,text="Result show Here",font=1,bg="green")
    lb1_result.grid(row=7,column=1,pady=20)
    
    
    lb2_result=Label(root1,text="Result show Here",font=1,bg="green")
    lb2_result.grid(row=8,column=1,pady=20)
    
 
    root1.resizable(False,False)
    
    root1.mainloop()
    
#-------------------------------chi2 ends----------------------------------------    
 

def t():
    
    
    root.destroy()
    root1=Tk()
    root1.geometry("600x500")
    root1.title("T-test Evaluation")
    
    def click():
        var11=list(map(int,var1.get().split(",")))
        
        var22=list(map(int,var2.get().split(",")))
        
        stats,p=ttest_ind(var11,var22)
        
        lb1_result.config(text="T-test value={0} and P-value={1}".format(round(stats,2),round(p,4))
        
        if p<(alpha):
                 lb2_result.config(text="{0}<{1}:".format(round(p,4),0.95)+
                             "Rject Null Hypothesis")
        else:
                 lb2_result.config(text="{0}>{1}:".format(round(p,4),0.95)+
                              "Fail to Reject Null Hypothesis") 
     
    lb_head=Label(root1,text="T-test Evaluation",bg="silver",fg="blue",font=1)
    lb_head.grid(row=0,column=0,padx=5)
    
    lb1_sample=Label(root1,text="Data_1 =",font=1,bg="gold"
                     ,padx=20)
    lb1_sample.grid(row=1,column=0,padx=5,pady=10)
    
    var1=StringVar()
    entry_sample_mean= Entry(root1,font=1,textvariable=var1, borderwidth=5,
                             width=30,cursor="arrow")
    entry_sample_mean.grid(row=1,column=1,padx=5)
    
    lb1_sample=Label(root1,text="Insert Column_1 elements seprated by comma",font=1,bg="silver"
                     ,fg="black"
                     ,padx=20)
    lb1_sample.grid(row=2,column=1,padx=5,pady=10)
    
   
    
    
    lb1_sample=Label(root1,text="Data_2 =",font=1,bg="gold"
                     ,padx=20)
    lb1_sample.grid(row=3,column=0,padx=5,pady=10)
    
    var2=StringVar()
    entry_sample_mean= Entry(root1,font=1,textvariable=var2, borderwidth=5,
                             width=30,cursor="arrow")
    entry_sample_mean.grid(row=3,column=1,padx=5)
    
    lb1_sample=Label(root1,text="Insert column_2 elements seprated by comma",
                     font=1,bg="silver"
                     ,fg="black"
                     ,padx=20)
    lb1_sample.grid(row=5,column=1,padx=5,pady=10)
    
       
    btn_calc= Button(root1,text="Press for Calculate",font=1,command=click)
    btn_calc.grid(row=6,column=1)
    
    lb1_result=Label(root1,text="Result show Here",font=1,bg="green")
    lb1_result.grid(row=7,column=1,pady=20)
    
    lb2_result=Label(root1,text="Result show Here",font=1,bg="green")
    lb2_result.grid(row=8,column=1,pady=20)
    
    
    root1.mainloop() 
    
  
 
def f():
  pass    
#--------------------------------main window-------------------------------------

def mainwindow():  
 
    root.geometry("300x300")
    root.title("Stats Calculator")
    #root.config(image=b"C:\Users\HP\OneDrive\Desktop\Shiva")
    lb_head=Label(root,text="*STATS CALCULATOR*",font="Lucida 19 bold",bg="black",
                  fg="white",
                  justify="left",pady=10).pack()
    #lb_head.grid(row=0,column=0)
    
    var=IntVar()
    rb1=Radiobutton(root,text="Z-score",font="Lucida 19 bold",variable=var,value=1,
                    command=z, fg="black",bg="white").pack(anchor="w")
    #rb1.config(padx=5,pady=5,bg="gold")
    #rb1.grid(row=1,column=0)
    
    
  
    rb2=Radiobutton(root,text="CHI2-value",font="Lucida 19 bold",variable=var,value=2
                    ,command=chi,
                    fg="black",bg="white").pack(anchor="w")
    #rb2.config(padx=10,pady=5,bg="gold")
    #rb2.grid(row=2,column=0)
    
 
    rb3=Radiobutton(root,text="F-Score",font="Lucida 19 bold",variable=var,value=3,
                    command=f,
                    fg="black",bg="white").pack(anchor="w")
    ##rb3.grid(row=3,column=0)
    
    rb4=Radiobutton(root,text="T-value",font="Lucida 19 bold",variable=var,value=4,
                    command=t,
                    fg="black",bg="white").pack(anchor="w")
    #rb4.config(padx=5,pady=5,bg="gold")
    #rb4.grid(row=4,column=0)
    
    root.resizable(False,False)
    
    
    root.mainloop()
    
#--------------------------main window ends----------------------------------------    
    
root =Tk()    
mainwindow()    #calling main function 
```
